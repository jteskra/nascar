```python
#Prepare training data
train = df[df['Race'] <= 143]
test = df[df['Race'] == 144]

train['Prev_Finish'] = train.groupby('Driver')['Finish'].shift(1)
mean_per_driver = train.groupby('Driver')['Prev_Finish'].transform('mean')
train['Prev_Finish'].fillna(mean_per_driver, inplace=True)


train['Prev_Laps'] = train.groupby('Driver')['Laps'].shift(1)
mean_per_driver = train.groupby('Driver')['Prev_Laps'].transform('mean')
train['Prev_Laps'].fillna(mean_per_driver, inplace=True)


train['Prev_Led'] = train.groupby('Driver')['Led'].shift(1)
mean_per_driver = train.groupby('Driver')['Prev_Led'].transform('mean')
train['Prev_Led'].fillna(mean_per_driver, inplace=True)

train['Prev_Pts'] = train.groupby('Driver')['Pts'].shift(1)
mean_per_driver = train.groupby('Driver')['Prev_Pts'].transform('mean')
train['Prev_Pts'].fillna(mean_per_driver, inplace=True)

train['Prev_Rating'] = train.groupby('Driver')['Rating'].shift(1)
mean_per_driver = train.groupby('Driver')['Prev_Rating'].transform('mean')
train['Prev_Rating'].fillna(mean_per_driver, inplace=True)

train['Prev_Status'] = train.groupby('Driver')['Status'].shift(1)
mean_per_driver = train.groupby('Driver')['Prev_Status'].transform('mean')
train['Prev_Status'].fillna(mean_per_driver, inplace=True)
train.fillna(0, inplace=True)

# Drop multiple columns by specifying their names in a list


train
```
